```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#PartA

env = gym.make("Taxi-v3")
state_size = env.observation_space.n
action_size = env.action_space.n

#Part B

q_table = np.zeros((state_size, action_size))

learning_rate = 0.1
discount_factor = 0.95
epsilon = 1.0
epsilon_decay = 0.9995
min_epsilon = 0.01
total_episodes = 2500

#Part C
rewards_per_episode = []

for episode in range(total_episodes):
    state, info = env.reset()
    terminated = False
    truncated = False
    total_reward = 0

    while not (terminated or truncated):

        if np.random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(q_table[state, :])


        next_state, reward, terminated, truncated, info = env.step(action)


        best_next_action = np.max(q_table[next_state, :])
        q_table[state, action] += learning_rate * (
            reward + (discount_factor * best_next_action) - q_table[state, action]
        )

        state = next_state
        total_reward += reward


    epsilon = max(min_epsilon, epsilon * epsilon_decay)
    rewards_per_episode.append(total_reward)

    if (episode + 1) % 500 == 0:
        print(f"Episode: {episode + 1} - Total Reward: {total_reward} - Epsilon: {epsilon:.4f}")


test_success = 0
for _ in range(100):
    state, info = env.reset()
    terminated = False
    while not terminated:
        action = np.argmax(q_table[state, :])
        state, reward, terminated, truncated, info = env.step(action)
        if reward == 20:
            test_success += 1

print(f"\nTraining Complete.")
print(f"Evaluation Success Rate: {(test_success/100)*100}%")


plt.figure(figsize=(14, 5))


plt.subplot(1, 2, 1)
plt.plot(rewards_per_episode, alpha=0.2, color='gray', label='Raw Reward')

moving_avg = np.convolve(rewards_per_episode, np.ones(50)/50, mode='valid')
plt.plot(moving_avg, color='blue', label='50-Ep Moving Avg')
plt.title("Learning Progress (Total Reward)")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.legend()


plt.subplot(1, 2, 2)
sns.heatmap(q_table[:50, :], cmap="magma")
plt.title("Q-Table Heatmap (States 0-50)")
plt.xlabel("Actions")
plt.ylabel("State Index")

plt.tight_layout()
plt.show()
```
